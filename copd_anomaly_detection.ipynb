## Step 6: Anomaly Detection
Identifying patients with unusual clinical readings using group-based 
anomaly detection.

Method: For each severity group, patients are flagged if their readings 
fall outside 2 standard deviations of their group mean. This approach 
is used rather than overall dataset averages because a low FEV1 is 
expected for a VERY SEVERE patient but unusual for a MILD patient.

Columns checked: FEV1, FVC, MWT1Best, CAT

```python
# calculate group mean and std for each column
# using transform so every patient has their group's stats next to them
columns_to_check = ['FEV1', 'FVC', 'MWT1Best', 'CAT']

for column in columns_to_check:
    df[f'{column}_group_mean'] = df.groupby('copd')[column].transform('mean')
    df[f'{column}_group_std'] = df.groupby('copd')[column].transform('std')
    
    # flag if outside 2 standard deviations of group mean
    df[f'{column}_anomaly'] = (
        (df[column] < df[f'{column}_group_mean'] - 2 * df[f'{column}_group_std']) |
        (df[column] > df[f'{column}_group_mean'] + 2 * df[f'{column}_group_std'])
    )

print("Anomalies found per measure:")
for column in columns_to_check:
    print(f"{column}: {df[f'{column}_anomaly'].sum()}")
```

```python
# combined flag - true if anomalous in any measure
df['any_anomaly'] = (
    df['FEV1_anomaly'] | 
    df['FVC_anomaly'] | 
    df['MWT1Best_anomaly'] | 
    df['CAT_anomaly']
)
print(f"Total flagged patients: {df['any_anomaly'].sum()}")
```

```python
# count how many measures flagged per patient
df['anomaly_count'] = (
    df['FEV1_anomaly'].astype(int) + 
    df['FVC_anomaly'].astype(int) + 
    df['MWT1Best_anomaly'].astype(int) + 
    df['CAT_anomaly'].astype(int)
)
print(df['anomaly_count'].value_counts())
```

```python
# priority level based on number of anomalies
def priority(count):
    if count == 0:
        return 'NORMAL'
    elif count >= 3:
        return 'CRITICAL'
    elif count == 2:
        return 'WARNING'
    else:
        return 'MONITOR'

df['priority_level'] = df['anomaly_count'].apply(priority)

print("Priority breakdown:")
print(df['priority_level'].value_counts())

print("\nTop priority patients:")
print(df[df['any_anomaly'] == True][['ID', 'copd', 'anomaly_count', 'priority_level']]
      .sort_values('anomaly_count', ascending=False))
```

### Anomaly Detection Findings

Group-based anomaly detection was run across 4 clinical measures 
(FEV1, FVC, MWT1Best, CAT) for 100 patients.

**Results:**
- 13 patients flagged as anomalous
- 1 patient identified as CRITICAL (ID 108 — flagged in 3 measures)
- 12 patients identified as MONITOR (flagged in 1 measure)

**Anomalies by Measure:**

  musculoskeletal comorbidities affecting walk test performance 
  independently of lung function
- FEV1: 3 anomalies
- FVC: 2 anomalies
